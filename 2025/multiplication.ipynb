# Naive Implementation (m*n)

```python
def naive_multiply(num1, num2):
    
    num1_str = str(num1)
    num2_str = str(num2)
    
    result = 0
    
    for i in range(0, len(num1_str)):
        for j in range(0, len(num2_str)):
            ans = (ord(num1_str[i])-ord('0'))*(ord(num2_str[j])-ord('0'))
            power_pos = 10**((len(num1_str)-1-i)+(len(num2_str)-1-j))
            
            result = result + ans*power_pos
    
    return result
```

# Kasturba Fast Multiplication Implementation

```python
def fastMultiplication(num1, num2):
    
    if num1 < 10 and num2 < 10:
        return num1*num2

    num1_len = len(str(num1))
    num2_len = len(str(num2))

    max_len = max(num1_len, num2_len)
    half_len = max_len//2

    divisor = 10**half_len
    res1 = num1//divisor
    rem1 = num1%divisor

    res2 = num2//divisor
    rem2 = num2%divisor

    first_part = fastMultiplication(res1, res2)
    second_part = fastMultiplication(rem1, rem2)
    third_part = fastMultiplication(res1 + rem1, res2 + rem2) - first_part - second_part

    return first_part*(10**(2*half_len)) + third_part*(10**half_len) + second_part
```

# Test Case 1

```python
start = time.time()
A=123
B=568
print("Multiplying", A, "by", B)
print("Naive Implementation Result: ", naive_multiply(A,B))
end = time.time()
print("Time Taken ", end-start)


print("---------------------------------------------------")

start = time.time()
print("Kasturba Implementation Result: ", fastMultiplication(A,B))
end = time.time()
print("Time Taken ", end-start)
```
